```python
dup_factor = {'Cell Phone (hand-Held)':'Cell Phone (hand-held)',
             'Drugs (illegal)':'Drugs (Illegal)',
             'Illnes':'Illness'}
```

```python
df_filter = df2.drop(['LATITUDE','LONGITUDE','LOCATION','ON STREET NAME','CROSS STREET NAME','OFF STREET NAME','UNIQUE KEY'], axis=1)
```

```python
for idx, row in df_filter.iterrows():
    for i in range(1,6):
        if row['CONTRIBUTING FACTOR VEHICLE '+str(i)] in dup_factor:
            temp = row['CONTRIBUTING FACTOR VEHICLE '+str(i)]
            df_filter.at[idx,'CONTRIBUTING FACTOR VEHICLE '+ str(i)] = dup_factor[temp]
        if row['VEHICLE TYPE CODE '+str(i)] in dup_vehicle:
            temp = row['VEHICLE TYPE CODE '+str(i)]
            df_filter.at[idx,'VEHICLE TYPE CODE '+ str(i)] = dup_vehicle[temp]
    
```

```python
df_filter.loc[:,"VEHICLE TYPE CODE 1"].value_counts()
```

```python
for idx, row in df_filter.iterrows():
    for i in range(1,6):
        if row['VEHICLE TYPE CODE '+str(i)] == 'BUS':
            df_filter.at[idx,'VEHICLE TYPE CODE '+ str(i)] = 'Bus'
    
```

```python
from collections import defaultdict
data = []
for idx, row in df_filter.iterrows():
    item = {}
    item['date'] = row['DATE']
    item['datetime'] = row['DATE'] + ' ' + row['TIME']
    item['borough'] = row['BOROUGH']
    item['zipcode'] = row['ZIP CODE']
    item['pedestrians_injured'] = row['NUMBER OF PEDESTRIANS INJURED']
    item['pedestrians_killed'] = row['NUMBER OF PEDESTRIANS KILLED']
    item['cyclist_injured'] = row['NUMBER OF CYCLIST INJURED']
    item['cyclist_killed'] = row['NUMBER OF CYCLIST KILLED']
    item['motorist_injured'] = row['NUMBER OF MOTORIST INJURED']
    item['motorist_killed'] = row['NUMBER OF MOTORIST KILLED']
    try:
        item['persons_injured'] = int(row['NUMBER OF PERSONS INJURED'])
    except:
        item['persons_injured'] = item['pedestrians_injured'] + item['cyclist_injured'] + item['motorist_injured']
    try:
        item['persons_killed'] = int(row['NUMBER OF PERSONS KILLED'])
    except:
        item['persons_killed'] = item['pedestrians_killed'] + item['cyclist_killed'] + item['motorist_killed']
    factors = {}
    vehicles = {}
    for i in range(1,6):
        factor = row['CONTRIBUTING FACTOR VEHICLE '+str(i)]
        if not pd.isna(factor):
            if factor in factors:
                factors[factor] += 1
            else:
                factors[factor] = 1
        vehicle = row['VEHICLE TYPE CODE '+str(i)]
        if not pd.isna(vehicle):
            if vehicle in vehicles:
                vehicles[vehicle] += 1
            else:
                vehicles[vehicle] = 1
    item['factor'] = factors
    item['vehicle'] = vehicles
    data.append(item)
```

```python
for idx, row in df_filter.iterrows():
    if pd.isna(row['NUMBER OF PERSONS INJURED']):
        df_filter.at[idx,'NUMBER OF PERSONS INJURED'] = row['NUMBER OF PEDESTRIANS INJURED'] + row['NUMBER OF CYCLIST INJURED'] + row['NUMBER OF MOTORIST INJURED']
    if pd.isna(row['NUMBER OF PERSONS KILLED']):
        df_filter.at[idx,'NUMBER OF PERSONS KILLED'] = row['NUMBER OF PEDESTRIANS KILLED'] + row['NUMBER OF CYCLIST KILLED'] + row['NUMBER OF MOTORIST KILLED']
    
    
```
